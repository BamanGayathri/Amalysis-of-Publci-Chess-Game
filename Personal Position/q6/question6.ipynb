```python
import pandas as pd

# Load the moves data
df = pd.read_csv('/content/Moves.csv')

# Manually create a dictionary or DataFrame for game results (example)
game_results = {
    1: 'Win',
    2: 'Loss',
    3: 'Draw',
    4: 'Win',
    # Add more games and their results here...
}

# Convert the game_results dictionary into a DataFrame
results_df = pd.DataFrame(list(game_results.items()), columns=['game_id', 'result'])

# Merge the results with the moves data based on game_id
df = df.merge(results_df, on='game_id', how='left')

# Now you can proceed to mark castling moves
df['is_castle'] = df['move'].apply(lambda x: 'King-side' if x == 'O-O' else ('Queen-side' if x == 'O-O-O' else 'No'))

# Filter out rows where the move is not a castling move
castling_df = df[df['is_castle'] != 'No']

# Group by game_id, is_castle and result, and count the occurrences
castling_outcomes = castling_df.groupby(['game_id', 'is_castle', 'result']).size().unstack(fill_value=0)

# Save to CSV for Tableau visualization
castling_outcomes.to_csv('CastlingImpactOnOutcomes.csv')
```
