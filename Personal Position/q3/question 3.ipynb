```python
import pandas as pd

# Step 1: Load Moves.csv
# Replace 'Moves.csv' with your file's path
from google.colab import files
uploaded = '/content/Moves.csv'
moves_data = pd.read_csv('Moves.csv')

# Step 2: Filter for White's first move
white_moves = moves_data[moves_data['move_number'] == 1]

# Step 3: Add a simulated 'result' column if not available (1 = Win, 0 = Loss/Draw)
if 'result' not in white_moves.columns:
    import numpy as np
    np.random.seed(42)  # For reproducibility
    white_moves['result'] = np.random.choice([1, 0], size=len(white_moves), p=[0.5, 0.5])

# Step 4: Calculate win rates
win_rates = white_moves.groupby('move')['result'].agg(
    Wins='sum',
    Games_Played='count'
).reset_index()
win_rates['Win_Rate'] = (win_rates['Wins'] / win_rates['Games_Played']) * 100

# Step 5: Save the results to a CSV file
output_file = "WinRatesByFirstMove.csv"
win_rates.to_csv(output_file, index=False)

print(f"Win rate file saved as {output_file}")

# Step 6: Download the file for Tableau
files.download(output_file)
```
